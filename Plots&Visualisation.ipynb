```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# import seaborn
import seaborn as sns
df = pd.read_csv ('bank.csv')
```

```python
g = sns.lmplot(y='age', # Variable 1.
               x='balance', # Variable 2.
               data=df, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlapping points.
g.set_ylabels("Age")
g.set_xlabels("Balance")
plt.title('Scatterplot: Age by Balance')
plt.show()
# how to make filter here?
```

```python
#who has more balance in the bank  #single or married?
# Setting the overall aesthetic.
sns.set(style="darkgrid")

# Use catplot (previous versions of seaborn use factorplot)
g = sns.factorplot(x="marital", y="balance", hue="housing", data=df, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Balamce")
g.set_xlabels("marital")
plt.title('Barplot: Deposit by Marital and Housing')
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set(style="whitegrid")

g = sns.pointplot(x="marital", y="balance", data=df, height=6, kind="strip", palette="pastel",ci=95,dodge=True,join=False)
plt.title('Pointplot: marital by deposit')
plt.show()
```
